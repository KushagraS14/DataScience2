```python
from surprise import Reader,Dataset,SVD  # SVD is the model,Dataset is the module from the surprise library which contains a very important function load_from_df
from surprise.model_selection import cross_validate
```

```python
#Cv as in 3 we need 3 folds of the data
#divided into 3 segments one ,three is training and second is testing ,second,third is training and last is testing and third is training and then others are testing
```

```python
#data can be divided into five folds we have better combinantions
```

```python
#we only work with 100K model to support quick runtime
data=Dataset.load_from_df(netflix_dataset[['Cust_Id','Movie_Id','Ratings']][:100000],reader)
```
