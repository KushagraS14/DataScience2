```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
train=new_data.iloc[:120]['maxpressurem_2']
```

```python
test=new_data.iloc[120:]
```

```python
model=ARIMA(train,order=(1,0,2))
```

```python
model_fit=model.fit()
```

```python
new_data['predicted']=model_fit.predict(start=len(train),end=len(train) + len(test)-1,dynamic=True)
new_data[['maxpressurem_2','predicted']].plot()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
new_data['predicted']=model_fit.predict(start=len(train),end=len(train) + len(test)-1,dynamic=True)
new_data[['maxpressurem_2','predicted']].plot()
```

```python
forecast=model_fit.forecast(steps=60)
plt.figure(figsize=(10,6))
plt.plot(new_data.index,new_data['maxpressurem_2'],label='Original Data')
```

```python
plt.plot(forecast.index,forecast,label='Forecast',linestyle='--')
plt.xlabel("Date")
plt.ylabel("maxpressurem_2")
plt.title("Forecasting Stock Prices for the next five years")
plt.legend()
```
