```python
# numerical data type is replaced by mean and object datatype like smoker, region is replaced by mode.
```

```python
for i in c :
  if takecare[i].dtype=='object' :
    takecare[i]=takecare[i].fillna(takecare[i].mode()[0])
  else:
    takecare[i]=takecare[i].fillna(takecare[i].mean())
```

```python
takecare.isnull().sum()
```

```python
col_list=[]
for  i in c:
  if takecare[i].dtype=='int64' or takecare[i].dtype=='float64':
    col_list.append(i)
    plt.boxplot(takecare[i])
    plt.xlabel(i)
    plt.ylabel('count')
    plt.show()
```

```python
q1=takecare['bmi'].quantile(0.25)
q3=takecare['bmi'].quantile(0.75)
iqr=q3-q1

takecare=takecare[(takecare['bmi']>q1-1.5*iqr) |(takecare['bmi']<q3+1.5*iqr)]


q1=takecare['past_consultations'].quantile(0.25)
q3=takecare['past_consultations'].quantile(0.75)
iqr=q3-q1

takecare=takecare[(takecare['past_consultations']>q1-1.5*iqr) |(takecare['past_consultations']<q3+1.5*iqr)]


q1=takecare['Hospital_expenditure'].quantile(0.25)
q3=takecare['Hospital_expenditure'].quantile(0.75)
iqr=q3-q1

takecare=takecare[(takecare['Hospital_expenditure']>q1-1.5*iqr) |(takecare['Hospital_expenditure']<q3+1.5*iqr)]



q1=takecare['NUmber_of_past_hospitalizations'].quantile(0.25)
q3=takecare['NUmber_of_past_hospitalizations'].quantile(0.75)
iqr=q3-q1

takecare=takecare[(takecare['NUmber_of_past_hospitalizations']>q1-1.5*iqr) |(takecare['NUmber_of_past_hospitalizations']<q3+1.5*iqr)]

q1=takecare['Anual_Salary'].quantile(0.25)
q3=takecare['Anual_Salary'].quantile(0.75)
iqr=q3-q1

takecare=takecare[(takecare['Anual_Salary']>q1-1.5*iqr) |(takecare['Anual_Salary']<q3+1.5*iqr)]



q1=takecare['charges'].quantile(0.25)
q3=takecare['charges'].quantile(0.75)
iqr=q3-q1

takecare=takecare[(takecare['charges']>q1-1.5*iqr) |(takecare['charges']<q3+1.5*iqr)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X=takecare.loc[:,['bmi','children','Anual_Salary']]
y=takecare.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
y_pred=lin_reg.predict(X_test)
```

```python
accuracy=r2_score(y_test,y_pred)
```

```python
accuracy
```

```python
sns.regplot(x=y_pred,y=y_test)
plt.xlabel('predicted_value')
plt.ylabel('actual_value')
plt.show()
```
