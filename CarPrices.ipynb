**Are the variance of wheelbase significantly different between hatchback and sedan    with gas fueltype**



*  Null Hypothesis:Variance of wheelbase  will be equal
*   Alternate hypothesis:Variance of wheelbase will not be equal




```python
new_df_hatchback=CarPrice[(CarPrice['carbody']==2) & (CarPrice['fueltype']==1)]
```

```python
new_df_hatchback.shape
```

```python
new_df_sedan=CarPrice[(CarPrice['carbody']==3) & (CarPrice['fueltype']==1)]
```

```python
new_df_sedan.shape
```

**Sampling**

```python
sample_hatchback=new_df_hatchback.sample(30,random_state=0)
sample_sedan=new_df_sedan.sample(30,random_state=0)
```

```python
a=sample_hatchback['wheelbase'].var()
b=sample_sedan['wheelbase'].var()
```

```python
print(f'Variance of drugY ,{a}')
print(f'Variance of drugA ,{b}')
```

```python
from scipy.stats import f
```

```python
F=a/b
```

```python
print(f"Ftest statistics",{F})
```

```python
dof_hatchback=len(sample_hatchback)-1
dof_sedan=len(sample_sedan)-1
```

```python
import scipy
p_value=scipy.stats.f.cdf(F,dof_hatchback,dof_sedan)
```

```python
if p_value>0.05:
  print('Failed to reject Null hypothesis,using this we can conclude there is enough evidences to conclude that Wheelbase will be equal')
else:
  print('Reject the null Hypotheisis there is not enough evidence to conclude that wheelbase will not be equal')
```
