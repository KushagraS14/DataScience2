**Are the variance of Na_to_K significantly different between drugA and drug Z    with high blood Pressure**



*  Null Hypothesis:Na_to_K will be equal
*   Alternate hypothesis:Na_to_K will not be equal




```python
new_df_drugY=drug[(drug['Drug']==0) & (drug['BP']==0)]
```

```python
new_df_drugY.shape
```

```python
new_df_drugA=drug[(drug['Drug']==2) & (drug['BP']==0)]
```

```python
new_df_drugA.shape
```

**Sampling**

```python
sample_drugY=new_df_drugY.sample(10,random_state=0)
sample_drugA=new_df_drugA.sample(10,random_state=0)
```

```python
a=sample_drugY['Na_to_K'].var()
b=sample_drugA['Na_to_K'].var()
```

```python
print(f'Variance of drugY ,{a}')
print(f'Variance of drugA ,{b}')
```

```python
from scipy.stats import f
```

```python
F=a/b
```

```python
print(f"Ftest statistics",{F})
```

```python
dof_drugY=len(sample_drugY)-1
dof_drugA=len(sample_drugA)-1
```

```python
import scipy
p_value=scipy.stats.f.cdf(F,dof_drugY,dof_drugA)
```

```python
if p_value>0.05:
  print('Failed to reject Null hypothesis,using this we can conclude there is enough evidences to conclude that NatoK will be equal for high BP')
else:
  print('Reject the null Hypotheisis there is not enough evidence to conclude that spends are same in both the NatoK will be equal for high BP')
```

**Based on the data that we have got we can use Decision tree classification model**

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
```

```python
X=drug.drop('Drug',axis=1)
y=drug['Drug']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
dtc=DecisionTreeClassifier(max_depth=5)
```

```python
plot_tree(dtc,filled=True) # filled=True is given to give the heatmap kind of analysis Darker colors have more weightage ,the lighter colors have leser weightage.overall in the tree which nodes end up having more importance
plt.show()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
```

When you created the model, how did you evaluate the model, and what are
some of the suggestions for improving the performance of the model?

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
