 Perform the following tasks for answering the remaining questions

 ● Remove the missing values from the dataset

```python
data=data.rename(columns={'annual_income': 'Annual Income'})
```

```python
data.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data.head()
```

```python
le=LabelEncoder()
```

```python
cols=list(data.columns)
```

```python
# for i in  cols:
#   data[i]=le.fit_transform(data[i])
```

```python
for i in data.columns:
    if data[i].dtype=='object':
        data[i]=le.fit_transform(data[i])
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('Annual Income',axis=1)
y=data['Annual Income']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

.What is the accuracy score of the above model?
 a. 0.60 to 0.70
 b. 0.40 to 0.60
 c. 0.70 to 0.85
 d. None of the above

```python
y_pred=log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#c)070 to 0.85
```
