#**It was observed that average clicks generated by Email is 550** Is it still the **same**?

```python
new_df=df.loc[(df['Campaign_Type']==1)]
```

```python
new_df.shape
```

```python
sample_df=new_df.sample(25,random_state=0)
```

```python
sample_df
```

```python
proposed_mean=550
```

```python
from scipy.stats import ttest_1samp
```

```python
tstatistics_value,p_value=ttest_1samp(sample_df['Clicks'],proposed_mean)
```

```python
if p_value>0.05:
  print("Failed to Reject Null Hypotheisis,there is  enough evidence to conclude that the average is  550")
else:
  print('Reject the Null Hypotheisis,there is  enough evidence to conclude that the average is  550')

```

```python
print(f'the Pvalue', {p_value})
```

** It was observed that the percentage of clicks genrated by camp1 that is more than 100 was 8% .Is it still the same? **

#Performing t test for 1 sample

```python
new_df=df[(df['Clicks']>100)]
```

```python
new_df.shape
```

```python
count=new_df['Campaign_Type'].value_counts()[0]
```

```python
count
```

```python
nobs=len(new_df['Campaign_Type'])
```

```python
new_df['Campaign_Type'].value_counts()/nobs
```

```python
p0=8
```

**Null Hypothesis :The proportion of Email CampaignType is 8%**

---


*   **Alternate Hypothesis: The proportion of Email Campaign Type is not 8% **

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
z_stats,p_value=proportions_ztest(count,nobs,p0)
```

```python
if p_value>0.05:
  print("Failed to Reject Null Hypotheisis,there is enough evidence to conclude that camp1 proportion is 8% ")
else:
  print('Reject the Null Hypotheisis,there is   not enough evidence to conclude that the camp1 proportion  is  8%')
```

**Are the variance of clicks significantly different between Email  and  Influencer among Men  **

```python
new_df_camp2=df[(df['Campaign_Type']==2) & (df['Target_Audience']==1)]
```

```python
new_df_camp2.shape
```

```python
new_df_camp2.shape
```

```python
new_df_camp1=df[(df['Campaign_Type']==1) & (df['Target_Audience']==1)]
```

**Sampling**

```python
sample_camp2=new_df_camp2.sample(500,random_state=0)
sample_camp1=new_df_camp1.sample(500,random_state=0)
```

```python
a=sample_camp2['Clicks'].var()
b=sample_camp1['Clicks'].var()
```

```python
print(f'Variance of camp2,{a}')
print(f'Variance of camp1,{b}')
```

```python
from scipy.stats import f
```

```python
F=a/b
```

```python
print(f"Ftest statistics",{F})
```

```python
dof_camp1=len(sample_camp1)-1
```

```python
dof_camp2=len(sample_camp2)-1
```

```python
import scipy
p_value=scipy.stats.f.cdf(F,dof_camp1,dof_camp2)
```

```python
if p_value>0.05:
  print('Failed to reject Null hypothesis,using this we can conclude there is enough evidences to conclude that clicks will be equal among Men')
else:
  print('Reject the null Hypotheisis there is not enough evidence to conclude that spends are same in both the clicks will be equal among Men ')
```
